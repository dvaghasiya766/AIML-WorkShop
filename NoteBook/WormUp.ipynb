## Numpy

### 1D Array

```python
# Create a NumPy array with 4 elements
arr = np.array([1, 2, 3, 4], dtype=np.int8)

# Print the elements of the array
print(arr)
```

```python
# Print the data type of the variable 'arr' (NumPy ndarray)
print('line: 02', type(arr))

# Print the number of dimensions of the array (1D array)
print('line: 05', arr.ndim)

# Print the shape of the array (rows, columns)
# For a 1D array, it returns (number_of_elements)
print('line: 09', arr.shape)
```

```python
# Print the total number of elements in the array
print('line: 02', arr.size)

# Print the size (in bytes) of one array element
print('line: 05', arr.itemsize)

# Print the total number of bytes consumed by the array
# Total bytes = size × itemsize
print('line: 09', arr.nbytes)
```

```python
# Print the Type of element of array
print('line: 02', arr.dtype)
```

### 3D Array

### Arithmetic Operations on Matrix

```python
arr1 = np.random.randint(0, 10, (3, 3))
print(arr1)
arr2 = np.random.randint(1, 10, (3, 3))
print(arr2)
```

```python
np.add(arr1, arr2)
```

```python
np.subtract(arr1, arr2)
```

```python
np.multiply(arr1, arr2)
```

```python
np.divide(arr1, arr2)
```
